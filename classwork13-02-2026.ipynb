```python
from skimage import io, color
import matplotlib.pyplot as plt
```

```python
url = "https://upload.wikimedia.org/wikipedia/commons/3/3f/Lichtenstein_img_processing_test.png"

image = io.imread(url)

plt.imshow(image)
plt.title("Original Image")
plt.axis("off")
```

```python
gray = color.rgb2gray(image)

plt.imshow(gray, cmap="gray")
plt.title("Grayscale")
plt.axis("off")
```

```python
from skimage.filters import sobel

edges = sobel(gray)

plt.imshow(edges, cmap="gray")
plt.title("Edges (Sobel)")
plt.axis("off")
```
